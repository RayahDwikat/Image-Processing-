```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img1 = cv2.imread('moon.jpg',1)
img2 = cv2.imread('magnitude.png',1)

img2 = cv2.resize(img2,(450,500))
img1 = cv2.resize(img1,(450,500))

dst = cv2.multiply(img1/255,img2/255)

res = np.hstack([dst])

plt.imshow(res)
plt.xticks([])
plt.yticks([])
plt.show()
```
